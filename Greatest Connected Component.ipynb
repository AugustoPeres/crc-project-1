```python
from itertools import count
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from importmarvel import create_marvel_graph
from colorspacious import cspace_converter
from collections import OrderedDict
import collections
```

```python
G=nx.read_graphml('connectcomp.graphml')
```

```python
print(nx.info(G))
```

```python
degfreq=nx.degree_histogram(G)
len(degfreq)
```

```python
deglist=list(range(0,1934))
```

```python
#DEGREE DISTRIBUTION
plt.plot(deglist,degfreq)
plt.show()
```

```python
#LOGLOG DEGREE DISTRIBUTION
plt.loglog(deglist,degfreq)
plt.show()
```

```python
#LOGLOG CUMULATIVE DEGREE DISTRIBUTION
cumdegfreq=[0]*1934
sum=0;

for i in range(0,1934):
    sum+=degfreq[i]
    cumdegfreq[i]=sum
cumdegfreq.reverse()
    
plt.loglog(deglist,cumdegfreq)
plt.show()
```

```python
#SAME AS BEFORE BUT CHANGING THE ORDER (OF THE COMULATIVE VALUES)
cumdegfreq.reverse()
plt.loglog(deglist,cumdegfreq)
plt.show()
```

```python
BG=nx.read_graphml('bipartitegraph.graphml')
BGlabels=nx.get_node_attributes(BG,'label')
```

```python
#DEGREE CENTRALITY
#GET THE HEROS WITH MOST CONNECTIONS (DEG CENTRALITY>= dg)

dg=0.17

degcent=nx.degree_centrality(G)

largestnodes=[]

for i in range(1,6486):
    if str(i) in degcent.keys():
        z= degcent[str(i)]
        if z>=dg:
            largestnodes.append([BGlabels[str(i)],z])
largestnodes
```

```python
len(largestnodes)
```

```python
#EIGENVECTOR CENTRALITY
#GET THE HEROS WITH MORE IMPORTANT CONNECTIONS (DEIGEN CENTRALITY>= Eg)

eg=0.09

eigencent=nx.eigenvector_centrality(G)

eigentopnodes=[]

for i in range(1,6486):
    if str(i) in eigencent.keys():
        z= eigencent[str(i)]
        if z>=eg:
            eigentopnodes.append([BGlabels[str(i)],z])
eigentopnodes
```

```python
len(eigentopnodes)
```

```python
#CLOSENESS CENTRALITY
#GET THE HEROS WITH MORE IMPORTANT CONNECTIONS (DEIGEN CENTRALITY>= Eg)

cc=0.09

closecent=nx.closeness_centrality(G)

closetopnodes=[]

for i in range(1,6486):
    if str(i) in closecent.keys():
        z= closecent[str(i)]
        if z>=cc:
            closetopnodes.append([BGlabels[str(i)],z])
closetopnodes
```
