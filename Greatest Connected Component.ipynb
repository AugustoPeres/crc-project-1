```python
#DEGREE DISTRIBUTION
plt.plot(deglist,degfreq)
plt.show()
```

```python
#LOGLOG DEGREE DISTRIBUTION
plt.loglog(deglist,degfreq)
plt.show()
```

```python
#LOGLOG CUMULATIVE DEGREE DISTRIBUTION
cumdegfreq=[0]*1934
sum=0;

for i in range(0,1934):
    sum+=degfreq[i]
    cumdegfreq[i]=sum
cumdegfreq.reverse()
    
plt.loglog(deglist,cumdegfreq)
plt.show()
```

```python
#SAME AS BEFORE BUT CHANGING THE ORDER (OF THE COMULATIVE VALUES)
cumdegfreq.reverse()
plt.loglog(deglist,cumdegfreq)
plt.show()
```
